```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Graficos Labos Ethernet

```python
dfEthernet = pd.read_csv("tablas/capturaEthernet_s1_table.csv", sep=";")
```

```python
#dfEntropyEther[0] -> entropy  ;   dfEntropyEther[1] -> max entropy
dfEntropyEther = dfEthernet["probability"][8:].reset_index(drop=True)
dfTableEthernet = dfEthernet[:8]
#print dfEntropyEther
#print dfTableEthernet

probability = dfTableEthernet["probability"]
symbols = dfTableEthernet["symbol"]

mpl_fig = plt.figure()
ax = mpl_fig.add_subplot(111)
ax.set_ylabel('Probabilidad')

plt.bar(range(len(probability)), probability, align='center')
plt.xticks(range(len(symbols)), symbols,rotation=45)

plt.plot([0, len(symbols)],[dfEntropyEther[0], dfEntropyEther[0]], 'k--', label='Entropia')
plt.text(sum(range(len(symbols)))/8, 1.73, 'Entropia')

#plt.plot([0,len(symbols)],[dfEntropyEther[1], dfEntropyEther[1]], 'k--', label='EntropiaMax')
#plt.text(sum(range(len(symbolsEther)))/5, 2.35, 'EntropiaMax')

plt.show()
```

```python
import json
import numpy as np
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap(projection='mill',lon_0=0)
map.drawcoastlines()

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')

with open('locations.json') as json_data:
    locations = json.load(json_data)

lat = [-34.5762]
lon = [-58.4872]
for entry in locations:
    lat.append(entry['latitude'])
    lon.append(entry['longitude'])
    
x, y = map(lon, lat)
map.plot(x, y, markersize=5, linewidth=1.5) 
#map.plot(currentLat, currentLon,linewidth=1.5,color='r')

plt.title('fuck u')
plt.show()
```
